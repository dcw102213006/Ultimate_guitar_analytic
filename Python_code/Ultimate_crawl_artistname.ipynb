請求網頁: 'https://www.ultimate-guitar.com/explore?genres[]=14&page=1' <br>
page:1~20  <br>
抓取Pop music的Artist name <br>
因為流行歌手的歌曲和弦比較不會那麼複雜

```python
from selenium import webdriver
driver = webdriver.Chrome() 

try:
    for page in range(1,21):

        #請求網址
        driver.get("https://www.ultimate-guitar.com/explore?genres[]=14&page="+str(page))
        print('Page:',page,':','https://www.ultimate-guitar.com/explore?genres[]=14&page='+str(page))
        html = driver.page_source       # get html
      
        soup1=BeautifulSoup(html,'lxml')
        row=soup1.find_all('a',class_='link-secondary _30ssZ _2-NA5')

        for i in row:
        #一個page大概50個row
        #for迴圈:50 個 element in row
        #每個element解析出Artist name與他的對應url      
        
            try:
                artist_name=i.text
                artist_url=i['href']

                bag={
                    'artist_name':artist_name,
                    'artist_url':artist_url,
                    'error':0
                }
            except:
                print('爬取失敗')
            finally:
                db=MongoConnect.mongoconnect()
                one_document=db.Ultimate_Guitar_Artist
                one_document.insert_one(bag)
                print('insert success:',artist_url)
                
        #爬一頁休息10秒
        time.sleep(10)   

except Exception as e:
    print(e)
    
```
