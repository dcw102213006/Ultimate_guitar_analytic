```python
from bs4 import BeautifulSoup
import requests
import re
import pymongo  
from pymongo import MongoClient
import pyodbc
import time
from datetime import datetime
import csv
import MongoConnect
from selenium import webdriver

headers = {
    'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.1 (KHTML, like Gecko) Chrome/22.0.1207.1 Safari/537.1"
}  
```

```python
db=MongoConnect.mongoconnect()
one_document=db.Ultimate_Guitar_EachTAB_Html_Beatles
list=one_document.find({'error':0})
data=[]
load=0
for i in list:
    load=load+1
    if(load % 100==0):print(load)
    data.append(i)
    
```

以下把Html Parse出Tab的資訊，包刮:歌手、譜名、版本、譜的所有和弦、譜的rating

```python
for i in data:
    try:
        tab_href=i['tab_href']
        web_html=i['web_html']
        soup1=BeautifulSoup(web_html,'lxml')
        chord_list=soup1.find_all('span','_3L0Da _3f5kI')
        chord_dict=[]
        for chord in chord_list:
            chord_dict.append(chord.text)
        tab_title=soup1.find('header',class_='_1UfHc').h1.text 
        artist=soup1.find('span',class_='_5eV9B').text
        version=soup1.find('a',class_='SCUIL').find('span',class_='_1Fap4').text
        rating=soup1.find('a',class_='SCUIL').find('div',class_='_1aOjb').text

        bag={
            'tab_href':tab_href,
            'web_html':web_html,
            'artist':artist,
            'tab_title':tab_title,
            'version':version,
            'rating':rating,
            'chord':chord_dict,
            'error':0

        }
    except Exception as e:
        print('Parse這張譜時有問題',e)
        bag={
            'tab_href':tab_href,
            'web_html':'',
            'artist':'',
            'tab_title':'',
            'version':'',
            'rating':'',
            'chord':'',
            'error':1

        }
    finally:###連接MONGO




        db=MongoConnect.mongoconnect()
        one_document=db.Ultimate_Guitar_Tab_data
        current_insert=one_document.find({'tab_href':tab_href}).count()
        
        if(current_insert==0):
            one_document.insert_one(bag)
            print('insert success:',tab_href)
        
        if(current_insert>0):print("已存在此頁面")
            
         
```

以下是測試用的程式

```python
from selenium import webdriver
driver = webdriver.Chrome() 

try:
    #請求網址
    driver.get("https://tabs.ultimate-guitar.com/tab/the_beatles/a_beginning_chords_2010143")


    #按下網頁上View all tabs按鈕
#     driver.find_element_by_class_name('_2eRTs').find_element_by_class_name('_38JBl').find_element_by_class_name('_3mzzu').click()
except Exception as e:
    #取得網頁html
    print('此tab只有一個Version',e)
finally:
    
    html = driver.page_source       # get html
    driver.close()

```

```python
soup1=BeautifulSoup(html,'lxml')
chord_list=soup1.find_all('span','_3L0Da _3f5kI')
chord_dict=[]
for chord in chord_list:
    chord_dict.append(chord.text)
tab_title=soup1.find('header',class_='_1UfHc').h1.text 
artist=soup1.find('span',class_='_5eV9B').text
version=soup1.find('a',class_='SCUIL').find('span',class_='_1Fap4').text
rating=soup1.find('a',class_='SCUIL').find('div',class_='_1aOjb').text

bag={
    
    'web_html':html,
    'artist':artist,
    'tab_title':tab_title,
    'version':version,
    'rating':rating,
    'chord':chord_dict,
    'error':0

}
```
