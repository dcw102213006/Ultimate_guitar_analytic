# 這支程式把Ultimate吉他譜資料表與Youtube下載的歌曲資料表做合併
合併後的資料表應該要有: <br>
每首歌曲的和弦，此歌曲擁有的吉他譜版本與對應的和弦 <br>
例:Song aaa有3張Ultimate吉他譜，則有Song aaa的和弦與Tab ver1~ver3的和弦 <br>

```python
#把原曲的和弦序列欄位拿出和弦

def retrie_song_chord(song_detector_chord):
    song_chord=[]
    for i in song_detector_chord:
        song_chord.append(i['ochord'])
    return song_chord
```

# 把每個Ultimate Guitar Tab的和弦序列，轉成跟原曲和弦序列同一個調
chord欄位從csv讀進來是字串，把它變回list 
用法: 
from ast import literal_eval ; literal_eval(song_train.chord[0])

```python
def chord_label_reduced(song_chord):
    chord_list=[]
    for i in song_chord:
        chord=cp.label_reduce(i,True).replace('::', ':')
        chord_list.append(chord)
    return chord_list  
```

先把兩個和弦欄位做label_reduce以保持一致的和弦標準格式

```python
df=song_train_merged.copy()
#literal_eval是把字串欄位變list，若已是list會有error

df['song_chord']=df['song_chord'].apply(literal_eval).apply(chord_label_reduced)
df['chord']=df['chord'].apply(literal_eval).apply(chord_label_reduced)
df2=df.copy()
df.head(2)
```

```python
#A:maj,maj是label_af_chords_cleaning的對象
def label_af_chords_cleaning(song_chord):
    chord_list=[]
    label_maj=[
        '7',
        'maj',
        'maj11',
        'maj13',
        'maj7',
        'maj7 V',
        'maj7#11',
        'maj7#5',
        'maj7b5',
        'maj7sus4',
        'majaj',
        'majaj7',
        'majaj9',
        'majajor',
        'majin',
        'major',
        'major7',
        'majn'
    ]
    label_min=[
        'min',
        'min7'  
    ]
    
    for i in song_chord:
        if i is  'N':chord_list.append(i)
        if i is not 'N':
            chord=i.split(':')[0]

            label=i.split(':')[1]


            if label in label_maj:
                label='maj'
            if label in label_min:
                label='min'

            chord_list.append(chord+':'+label)
    return chord_list  
```

```python
df['chord']=df['chord'].apply(label_af_chords_cleaning)
df['song_chord']=df['song_chord'].apply(label_af_chords_cleaning)
df['chord'].head(2)
```
