# Youtube爬蟲

把要下載的音檔的youtube影片url爬完

```python
driver = webdriver.Chrome()     # 打開Chrome 瀏覽器，webdriver.Chrome('下載的Chrome driver路徑.exe')


for song in song_name_to_crawl:
    
    url='https://www.youtube.com/results?search_query='+song

    try:
        driver.get(url)#請求網址
        page_html = driver.page_source
        soup1=BeautifulSoup(page_html,'lxml')
        song_yt_url=soup1.find('a',id='video-title')['href']
        song_yt_title=soup1.find('a',id='video-title')['title']
        bag={
            'song_yt_url':'https://www.youtube.com'+song_yt_url,
            'song_yt_title':song_yt_title,
            'song_name':song,
            'error':0
        }
    except Exception as e:
        bag={
            'song_name':song,
            'error':1
        }
    finally:
        db=MongoConnect.mongoconnect()
        one_document=db.Youtube_songtodownload
        one_document.insert_one(bag)
        print('insert success:',song)
        time.sleep(10)   
```

# Youtube Download 
用每個youtube影片的url下載音檔

```python
db=MongoConnect.mongoconnect()
one_document=db.Youtube_songtodownload
list=one_document.find({'error':0})
download_list=[]
load=0
for i in list:
    load=load+1
    if(load % 100==0):print(load)
    download_list.append(i)
```

```python
pd.DataFrame(download_list).head()
```

```python
for i in download_list:
    print(i['song_yt_url'])
```

```python

for i  in download_list:
    try:
        link = i['song_yt_url']
        song_name=i['song_name']
        yt = YouTube(link)
        stream = yt.streams.filter(file_extension='mp4', res='360p').first()
        stream.download("./youtube_song_download")
        oldfilename='./youtube_song_download/'+yt.streams.first().default_filename
        oldfilename
        os.rename(oldfilename, './youtube_song_download/'+song_name+'.mp4')
        bag={
            'song_yt_url':link,
            'song_name':song_name,
            'error':0
        }
    except Exception as e:
        print(e)
        bag={
            'song_yt_url':link,
            'song_name':song_name,
            'error':1
        }
    finally:
        db=MongoConnect.mongoconnect()
        one_document=db.Youtube_has_downloaded
        one_document.insert_one(bag)
        print('insert success:',song_name)
        time.sleep(10)   
```
