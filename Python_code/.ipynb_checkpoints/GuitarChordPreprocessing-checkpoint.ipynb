```python
chord_reduce_list=[]
chord_reduce_eachrow=[]
for row in range(len(song_train)):
    
    chord_list=literal_eval(song_train.iloc[row].chord)
    for chord in chord_list:
        chord_reduce_eachrow.append(cp.label_reduce(chord))
    chord_reduce_list.append(chord_reduce_eachrow) 
    chord_reduce_eachrow=[]
    
print('song_train長度',len(song_train))
print('和弦進行資料長度',len(chord_reduce_list))
```

```python
chord_reduce_list[1]
```

```python
c_key_chord_progression_list=[]
for chord_progression in chord_reduce_list:
    estimate_key=cp.key_estimate(chord_progression)['estimate_key']
    c_key_chord_progression=cp.smooth_key(chord_progression,estimate_key,cp.C)
    c_key_chord_progression_list.append(c_key_chord_progression)
```

```python
for chord_progression in c_key_chord_progression_list:
    print(cp.key_estimate(chord_progression))
```
